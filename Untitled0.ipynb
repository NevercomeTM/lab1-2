```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
infa = pd.read_csv('train.csv')
infa
```

```python
import re, string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
def textprocess(text):
    text = text.lower()
    for i in string.punctuation:
        text = text.replace(i, ' ')
    tokens = re.split(' +', text)
    mass = []
    for i in tokens:
        if i not in stopwords.words('english'):
            mass.append(i)
    return mass
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.neighbors import KNeighborsClassifier
vectorer = TfidfVectorizer(max_df = 500, min_df = 10)
texts = [' '.join(textprocess(i)) for i in infa['sms']]
infa2 = infa
infa2['sms'] = texts
infa2
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(infa2['sms'], infa2['label'], test_size = 0.2)
xtrain = vectorer.fit_transform(xtrain)
xtest = vectorer.transform(xtest)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
knn = KNeighborsClassifier()
knn.fit(xtrain,ytrain)
prediction = knn.predict(xtest)
print(confusion_matrix(prediction, ytest))
print(classification_report(prediction, ytest))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(xtrain.toarray(), ytrain)
prediction = gnb.predict(xtest.toarray())
print(confusion_matrix(prediction, ytest))
print(classification_report(prediction, ytest))
```
